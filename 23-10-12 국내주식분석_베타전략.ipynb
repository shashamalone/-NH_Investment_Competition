## 전략1. 가치주 전략으로 국내주식데이터 분석하기

#### 1. 베타 
- 베타(Beta) : 개별 주식이 전체 주식시장의 변동에 반응하는 정도를 나타낸 값 

'주식시장의 움직임'은 개별 주식의 수익률에 가장 크게 영향을 주는 '팩터'이다
아무리 좋은 주식도 주식시장이 폭락한다면 내려가고, 나쁜 주식도 주식시장이 상승한다면 같이 상승한다.

베타 0.5 -> 주식시장의 움직임에 절반수준으로 움직인다
베타 1 -> 주식시장과 움직임이 정확히 같다.
베타 1.5 -> 주식시장 움직임이 +1% 일때, 개별 주식 수익률이 1.5%움직임을 보인다는 뜻이다.

=> 결론 : **상승장에서는 베타가 큰 주식의 수익률이 더 좋음.  하락장에서는 베타가 낮은 주식의 수익률이 더 좋다.**

 beta
| 베타 | 주식시장이 +1% 일 경우 | 주식시장이 -1% 일 경우 | 
| --- | --- | --- |
| 0.5 | +0.5% | -0.5% |
| 1.0 | +1.0% | -1.0% |
| 1.5 | +1.5% | -1.5% |


- 주식 베타는 통계의 회귀분석모형에서 기울기를 나타내는 베타와 의미가  같음
- 


- 회귀분석모형 : y = ax +b => 자산가격결정모형 
-  
- 

=> 개별 주식의 수익률은 무위험 수익률과 주식시장 수익률움직임에 얼마나 민감하게 반응하는가 의 합으로 알 수 있음

Ri : 개별주식의 수익률

Rf : 무위험 자산의 수익률

Bi : 개별주식의 베타

Rm : 주식시장의 수익률(전체 주식지상의 수익률에서 무위험 자산의 수익률을빼야함)




#### 1.1 베타 계산하기
- 주식시장에 대한 대용치 - KOSPI 지수
    - 코스피(한국종합주가지수) : 유가증가증권시장에서 거래되는 주식가격변동을 종합정리해서 만든 지수
    - 대기업들의 주식이 거래되는 제 1시장
    - 코스피에 상장되기 위해서는 300억원 이상의 자기자본/상장주식수 100만주이상/일반주주 700명이상 등
    - 코스닥(): 중소기업과 벤처기업을 위한 증권시장
    - https://www.youtube.com/watch?v=oyBiNC7IEKA
- 개별주식 - 전통적 고베타주인 증권주 이용

```python
#package
import yfinance as yf
import pandas as pd
```

```python
tickers = ['^KS11', '039490.KS']  #티커정보 - 코스피지수, 키움증권티커(한국종목 티커.KS, 티커.KQ)
# KOSPI 코드(^KS11)와 증권주인 키움증권의 티커(039490.KS)를 입력한다

#주가 다운로드
all_data = {} #주가 : key부분에는 kicer(코스피, 증권주 종목), value부분에 데이터 다운
for ticker in tickers:
    all_data[ticker] = yf.download(ticker,
                                   start="2022-08-31",
                                   end='2023-08-31') 

#주가를 데이터프레임으로 만들기  
## 종가에 해당하는 부분만 뽑아서 데이터프레임으로 치환
prices = pd.DataFrame({tic: data['Close'] for tic, data in all_data.items()}) 
## '주가다운 -> 종가를 데이터프레임으로 만ㄷ름 -> 수익률계산'과정에 관용적으로 쓰이는 코드
```

```python
#prices 데이터 프레임
# 열에는 ticker가 들어감 - 코스피 & 종목, 행에는 - 날짜, 값에는 - 종가에 해당하는 부분만 선택됨
prices
```

```python
# 수익률 계산
# pct_change() 함수를 통해 수익률을 계산한 후, NA 데이터는 삭제한다.
ret = prices.pct_change().dropna() # 맨 위값 - 이전일의 주가 x , 제거해줌
ret
```

수익률을 구했으니 증권주를 대상으로 베타를 구하기 위한 회귀분석을 실시한다. 
자산가격결정모형에서 편의를 위해 무위험 수익률인 Rf를 0으로 가정하면, Ri = Bi *Rm의 형태로 나타냄
이 중 Rm는 독립변수인 주식시장의 수익률을 의미하고, Ri는 종속변수인 개별 주식의 수익률을 의미한다.

1. 먼저 알파를 계산하기 위헤 intercept(절편)에 해당하는 열에 1을 입력한다.
2. 종속변수에는 증권주, 독립변수에는 KOSPI 지수 수익률 및 절편을 입력한다.
3. statsmodels 패키지의 OLS() 함수를 통해 손쉽게 선형회귀분석을 실시할 수 있으며, 회귀분석의 결과를 reg 변수에 저장한다.

```python
# pip install statsmodels.api
```

```python
import statsmodels.api as sm

ret['intercept'] = 1 #절편 : 수익률table에 1값 입력
reg = sm.OLS(ret[['039490.KS']], ret[['^KS11', 'intercept']]).fit() #OLS-선형회귀분석
# y축에는 개별주식 입력, x축에는 코스피 수익률 + intercept
```

```python
#결과도출
reg.summary()

# Dep. Variable:	039490.KS : 종속변수가 개별주식
# Adj. R-squared: 설명력 39% - 높은 편
# coef 베타 1.5148 -> 증권주의 경우 코스피의 움직임에 1.51배 반응한다는 것 - 고베타주라는 뜻
#t-통계값이 절댓값기준 1보다 크면 통계적으로 유의 
# -> 12.763이므로 개별주식의 수익률을 설명하는데 주식시장의 수익률이 굉장히 유의미한 설명변수라는 뜻
#알파(초과수익률)에 해당하는 intercept의 coef값이 0.0007, T-value도 유의하지 않음
# -> 증권주의 수익률응ㄴ 주식시장에 대한 노출도인 베타를 제외하고 나면 초과수익이 없다고 볼 수 있음
```

#### 2.밸류 전략(가치주 전략)을 이용한 종목 선정

- 가치주 효과 :
**내재 가치 대비 낮은 가격의 주식(저PER, 저PBR)**이 내재 가치 대비 비싼 주식(고PER, 고PBR)보다 수익률이 높은 현상

가치주 효과가 발생하는 원인은 사람들이 가치주를 기피하고 성장주를 선호하기 때문 -> 사람들이 기피한 주식이 가치주가 된것

가치주는 일반적으로 차입비율이 높고, 수익의 변동성이 크며, 경기가 좋지 않을 때 더 위험
-> 사람들은 위험한 주식에 필요 이상으로 과민반응을 보임, 그로인해 주가가 하락하고 가치주가 된다
반면 인간은 익숙한 것을 안전하다고 착각하는 경향이 있어서 최근 성과가 좋은 주식은 여러 매체를 통해 접하기 쉽고, 이런 주식을 안전하다고 착각해 많은 사람들이 매수한다

그로 인해 주가가 상승하고 고평가 주가 된다

투자자들의 확증 편향에 의해 위험하다고 생각되는 가치주는 더욱 기피하고 안전하다고 생각되는 성장주는 상승할 것같은 이유만 찾아 더욱 선호하게 된다.
가치주가 생각보다 위험하지 않다는 것, 성장주가 너무많이상승해 안전하지 않다는 것을 깨달으면 주가는 원래 수준으로 회귀한다 -> 이로 인해 가치주 효과가 발생한다

#### PER과 PBR이 낮은 종목 찾아내기
1. PER 내에서의 순위, PBR 내에서의 순위를 구한다.
- rank() 함수를 통해 PER와 PBR 열의 순위를 구하며, axis = 0을 입력하여 순위는 열 방향으로 구한다. 
2. sum() 함수 내에 axis = 1를 통해 위에서 구한 순위를 행 방향으로 값을 더하며, skipna = False를 통해 NA가 있는 종목은 제외한다. 그 후 다시 rank() 함수를 통해 순위의 합 기준으로 다시 순위를 구한다.
3. value_sum <= 5000, 즉 순위가 낮은 20 종목을 선택한다. 이는 PER과 PBR가 낮은 종목이라고 볼 수 있다.

```python
value_rank = tidy_fin[['PER', 'PBR']].rank(axis = 0)
value_sum = value_rank.sum(axis = 1, skipna = False).rank()
tidy_fin.loc[value_sum <= 5000 , ['종목코드', '종목명', 'PER', 'PBR']].drop_duplicates(['종목코드']).head(20)
```

#### 여러 지표 결합하기

가치지표에 해당하는 모든 지표들을 고려한 밸류 포트폴리오를 만들어보도록하자.
'PER', 'PBR','BPS', 'EPS','DIV','DPS'와 같은 가치지표를 고려하여 지표별 상관관계를 살펴본다.

1. 가치 지표들은 아래와 같은 상관관계를 가지고 있다.

    1.1.
    PER (Price-to-Earnings Ratio) - 값이 낮을수록 가치주에 해당
    PBR (Price-to-Book Ratio) - 값이 낮을수록 가치주에 해당

    1.2. BPS, EPS, DIV, DPS의 경우 값이 높을수록 배당수익률이 높은  가치주에 해당한다.  따라서 BPS, EPS, DIV, DPS에는 역수를 취해 순서를 맞춰준다.
    BPS (Book Value Per Share) - 값이 높을수록 가치주에 해당
    EPS (Earnings Per Share) - 값이 높을수록 가치주에 해당
    DIV (Dividend Yield) - 값이 높을수록 가치주에 해당
    DPS (Dividends Per Share) - 값이 높을수록 가치주에 해당

 
2. 각 지표의 순위를 열 방향으로 구한다.
3. heatmap() 함수를 통해 지표 별 상관관계를 히트맵으로 나타낸다.

```python
import matplotlib.pyplot as plt
import seaborn as sns

value_list_copy = tidy_all.copy()
value_list_copy['BPS'] = 1 / value_list_copy['BPS']
value_list_copy['EPS'] = 1 / value_list_copy['EPS']
value_list_copy['DIV'] = 1 / value_list_copy['DIV']
value_list_copy['DPS'] = 1 / value_list_copy['DPS']
value_list_copy = value_list_copy[['PER', 'PBR','BPS', 'EPS','DIV','DPS']]
value_rank_all = value_list_copy.rank(axis=0)
mask = np.triu(value_rank_all.corr())

mask = np.triu(value_rank_all.corr())
fig, ax = plt.subplots(figsize=(10, 6))
sns.heatmap(value_rank_all.corr(),
            annot=True,
            mask=mask,
            annot_kws={"size": 16},
            vmin=0,
            vmax=1,
            center=0.5,
            cmap='coolwarm',
            square=True)
ax.invert_yaxis()
plt.show()
```

DPS와 BPS,EPS,DIV/ PER과 PER, BPS와 EPS가 서로 높은 상관관계를 나타냈다.

DIV는 PER과 PBR과도 0.61,0.58의 비교적 높은 상관관계를 나타냈다.


얖비슷한 가치지표임에도 불구하고 서로 간의 상관관계가 꽤 낮은 지표도 있다. 따라서 지표를 통합적으로 고려하면 분산효과를 기대할 수도 있다.

## K-RATIO

```python
import statsmodels.api as sm
import numpy as np

ret = df.pct_change().iloc[1:]
ret_cum = np.log(1 + ret).cumsum()

x = np.array(range(len(ret)))
y = ret_cum.iloc[:, 0].values
```

```python
df
```

```python
reg = sm.OLS(y, x).fit()
reg.summary()
```

```python
x = np.array(range(len(ret)))
k_ratio = {}

for i in range(0, 200):

    ticker = data_bind.loc[i, '종목코드']

    try:
        y = ret_cum.loc[:, df.columns == ticker]
        reg = sm.OLS(y, x).fit()
        res = float(reg.params / reg.bse)
    except:
        res = np.nan

    k_ratio[ticker] = res

k_ratio_bind = pd.DataFrame.from_dict(k_ratio, orient='index').reset_index()
k_ratio_bind.columns = ['종목코드', 'K_ratio']

k_ratio_bind
```

```python
data_bind = data_bind.merge(k_ratio_bind, how='left', on='종목코드')
k_ratio_rank = data_bind['K_ratio'].rank(axis=0, ascending=False)
data_bind[k_ratio_rank <= 20]
```
